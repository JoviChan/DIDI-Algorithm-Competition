```python
import os
from dateutil.parser import parse as dateutil_parse
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

PATH = 'season_1/'
CLEAN_PATH = PATH+'clean/'

%matplotlib inline
```

```python
gaps1 = pd.read_csv(CLEAN_PATH+'train_gap.csv',index_col=0)
gaps2 = pd.read_csv(CLEAN_PATH+'test_gap.csv',index_col=0)
```

# find the most matched day from triaining data, use its gap as answer

```python
best_choice = np.zeros((5,2))
best_choice[0]
```

```python
for i in range(len(gaps2.columns)):
    for j in range(len(gaps1.columns)):
        j_theta = 0
        for k in range(len(gaps2)):
            if(gaps2.ix(k, i) != 0):
                s = gaps1.ix[k, j] - gaps2.ix[k, i]
                j_theta += s*s
        if(j_theta < best_choice[i][1] or best_choice[i][1] == 0):
            best_choice[i][0] = j
            best_choice[i][1] = j_theta
```

```python
best_choice
```

```python
asked_gap = pd.read_table(PATH + 'test_set_1/read_me_1.txt', header=None)
```

```python
asked_gap['answer'] = np.zeros((len(asked_gap)))
```

```python
asked_gap.columns = ['year','month','day','piece','answer']
```

```python
asked_gap
```

```python
asked_gap.loc[0, 'piece']
```

# generate answer

```python
for i in range(len(asked_gap)):
    asked_gap.loc[i, 'answer'] = gaps1.ix[asked_gap.ix[i, 'piece'], int(best_choice[(asked_gap.ix[i, 'day'] - 22) / 2][0])]
    if(asked_gap.loc[i, 'answer'] < 300):
        asked_gap.loc[i, 'answer'] = 0
```

```python
asked_gap
```

```python
asked_gap.to_csv(PATH+'clean/asked_gap.csv')
```

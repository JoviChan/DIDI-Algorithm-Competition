```python
def generateNaiveOrder(path = PATH+'training_data/order_data/'):
    order_paths = filter(lambda x: len(x.split('.'))==1,os.listdir(path))
    naive_order = pd.DataFrame()
    for order_path in order_paths:
        tmp = pd.read_table(path+order_path, header=None,index_col=0)
        tmp.index.name = 'order_id'
        tmp.columns = ['driver_id','passenger_id','start_district_hash','dest_district_hash','Price','Time']
        tmp['request'] = tmp['driver_id'].fillna(-1).map(lambda x: False if x==-1 else True)
        naive_order = pd.concat([naive_order,tmp[['request','start_district_hash','Price','Time']]])
    naive_order.index = range(len(naive_order.index))
    naive_order['Time'] = naive_order['Time'].map(lambda x: dateutil_parse(x))
    return naive_order
```

```python
def generateWeather(path = PATH+'training_data/weather_data/'):
    weather_paths = filter(lambda x: len(x.split('.'))==1,os.listdir(path))
    weather = pd.DataFrame()
    for wpath in weather_paths:
        tmp = pd.read_table(path+wpath, header=None,index_col=0,parse_dates=True)
        tmp.index.name = 'datatime'
        tmp.columns = ['weather','temperature','pm25']
        weather = pd.concat([weather,tmp])
    return weather
```

```python
generateWeather(path = PATH+'training_data/weather_data/').to_pickle(PATH+'clean/train_weather.pickle')
pd.read_pickle(PATH+'clean/train_weather.pickle').info()
```

```python
def generateTraffic(path = PATH+'test_set_1/traffic_data/'):
    def parse_level(s, level_flag):
        level, num = s.split(':')
        if int(level) != level_flag:
            raise ValueError('%s don\'t match level of %s' % (level,level_flag))
        return int(num)
    traffic_paths = filter(lambda x: len(x.split('.'))==1,os.listdir(path))
    traffic = pd.DataFrame()
    for tpath in traffic_paths:
        tmp = pd.read_table(path+tpath, header=None,index_col=0)
        tmp.index.name = 'start_district_hash'
        levels = ['level_'+str(i) for i in range(1,5)]
        for i, level in enumerate(levels):
            tmp[level] = tmp[i+1].map(lambda x: parse_level(x,i+1))
        traffic = pd.concat([traffic,tmp[levels+[5]]])
    traffic['datetime'] = traffic[5].map(lambda x: dateutil_parse(x))
    return traffic.drop(5, axis=1)
```

```python
generateTraffic(path = PATH+'test_set_1/traffic_data/').to_pickle(PATH+'clean/test_traffic.pickle')
pd.read_pickle(PATH+'clean/test_traffic.pickle').info()
```
